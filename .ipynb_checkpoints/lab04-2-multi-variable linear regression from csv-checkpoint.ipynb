````python
import tensorflow as tf
import numpy as np
tf.set_random_seed(777)
````

````python
xy = np.loadtxt('data/data-01-test-score.csv', delimiter=',',
                dtype=np.float32)
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]

print(x_data.shape, x_data, len(x_data))
print(y_data.shape, y_data)
````

````python
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])
````

````python
W = tf.Variable(tf.random_normal([3, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
````

````python
hypothesis = tf.matmul(X, W) + b
````

````python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
````

````python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
````

````python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001): 
    cost_val, hy_val, _ = sess.run(
        [cost, hypothesis, train],
        feed_dict={X: x_data, Y: y_data})
    if step % 10 == 0:
        print(step, "Cost: ",cost_val,
                     "\nPrediction:\n", hy_val)
````

````python
print("Your score will be ", sess.run(hypothesis,
                                     feed_dict={X: [[100, 70, 101]]}))
print("Other scores will be ", sess.run(hypothesis,
                                       feed_dict={X: [[60, 70, 110], [73, 80, 75]]}))
````

# Queue Runners

학습시킬 파일의 크기가 크거나 파일의 갯수가 많을 때 메모리에 부하가 많이 걸림
이에 대체가능한 것이 Tensorflow의 Queue Runner

파일들을 입력받음 -> **Queue**에다가 쌓음 -> **Reader**로 읽음 -> **Decode**를 함 -> **Queue**에 다시 쌓음

입력받은 파일들이 모두 필요한 것은 아니니 필요 시 Batch만큼 꺼내서 사용한다.

```python
filename_queue = tf.train.string_input_producer(
    ['data-01-test-score.csv', 'data-02-test-score.csv', ...],
    shuffle=False, name='filename_queue')
    
reader = tf.TextLineReader()
key, value = reader.read(filename_queue)

# Data Type 정할 때 사용하는 변수
record_defaults = [[0.], [0.], [0.], [0.]]
xy = tf.decode_csv(value, record_defaults=record_defaults)
```

## tf.train.batch

```python
train_x_batch, train_y_batch = \
    tf.train.batch([xy[0:1], xy[-1:]], batch_size=10)
    # arg1 = x_data, arg2 = y_data, arg3 = 한번에 가져올 갯수
    
sess = tf.Session()

coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess=sess, coord=coord)

for step in range(2001):
    x_batch, y_batch = sess.run([train_x_batch, train_y_batch])
    
coord.request_stop()
coord.join(threads)
```

````python
import tensorflow as tf
````

````python
filename_queue = tf.train.string_input_producer(
    ['data/data-01-test-score.csv'], shuffle=False, name='filename_queue')
````

````python
reader = tf.TextLineReader()
key, value = reader.read(filename_queue)
````

````python
record_defaults = [[0.], [0.], [0.], [0.]]
xy = tf.decode_csv(value, record_defaults=record_defaults)
````

````python
train_x_batch, train_y_batch = \
    tf.train.batch([xy[0:-1], xy[-1:]], batch_size=6)
````

````python
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])
````

````python
W = tf.Variable(tf.random_normal([3, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
````

````python
hypothesis = tf.matmul(X, W) + b
````

````python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
````

````python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
````

````python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess=sess, coord=coord)

for step in range(2001): 
    x_batch, y_batch = sess.run([train_x_batch, train_y_batch])
    cost_val, hy_val, _ = sess.run(
        [cost, hypothesis, train],
        feed_dict={X: x_batch, Y: y_batch})
    if step % 10 == 0:
        print(step, "Cost: ",cost_val,
                     "Prediction: ", hy_val)
        
coord.request_stop()
coord.join(threads)
````

````python
print("Your score will be ", sess.run(hypothesis,
                                     feed_dict={X: [[100, 70, 101]]}))
print("Other scores will be ", sess.run(hypothesis,
                                       feed_dict={X: [[60, 70, 110], [73, 80, 75]]}))
````
